* 시간초과 ㅠㅠ

```python
def ran(orders):
    rang = []
    count = 0
    
    for order in orders:
        for alpha in order:
            if alpha not in rang:
                rang.append(alpha)
            
        if (count < len(order)):
            count = len(order)
        
    return rang, count
        
def cal (orders, answers, count, rang, alpha, k):
    i = 0
    
    if (k <= 0):
        i = len(orders)
        
        for order in orders:
            for a in alpha:
                if a not in order:
                    i -= 1
                    break;
        
        answers = []
        answers.append(alpha)
        return i, answers
    
    for a in rang:
        i += 1
        c, ans = cal(orders, answers, count, rang[i:], alpha + a, k - 1)
        
        if (count < c):
            answers = ans
            count = c
        elif (count == c):
            answers += ans
            answers = set(answers)
            answers = list(answers)
    
    return count, answers

def solution(orders, course):
    answer = []
    
    rang, count = ran(orders)
    rang.sort()
    
    for k in course:
        if (count < k):
            break
        
        c, ans = cal(orders, [], 0, rang, "", k)
        
        if (c >= 2):
            answer += ans
    
    answer.sort()
    
    return answer
```
